```python
import yfinance as yf
import csv
from datetime import datetime
import time

#file to store historic stock prices
csv_file = "stocks_prices.csv"

#stocks to track
stocks = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA", "NVDA", "META", "BRK-B", "JPM", "V"]

#fetch every hour
fetch_interval = 3600

while True:
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    row = [timestamp]
    
    print(f"Fetching stock prices at {timestamp}...")
    for symbol in stocks:
        try:
            price = yf.Ticker(symbol).info["regularMarketPrice"]
            print(f"{symbol}: {price}")
            row.append(price)
        except Exception as e:
            print(f"Error fetching {symbol}: {e}")
            row.append(None)

    #save
    with open(csv_file, mode="a", newline="") as f:
        writer = csv.writer(f)
        if f.tell() == 0:
            #write header for empty file
            writer.writerow(["Timestamp"] + stocks)
        writer.writerow(row)

    print(f"Data saved. Waiting {fetch_interval / 3600} hours until next fetch...\n")
    time.sleep(fetch_interval)
```
